```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')


al = pd.read_csv('S:/Hackathons/EY/results/scaled_dist_data.csv')
```

```python
al.columns
```

```python
prep_indexed = al[['active', 'density', 'urbans', 'full_immunes', 'elderlies','hh_more_than_4', 'cold_chain_points',
                   'storage_cap_lits']]
prep_indexed.head()
```

```python
prep_indexed['full_immunes']  = 1-prep_indexed['full_immunes'] 
prep_indexed['cold_chain_points']  = 1-prep_indexed['cold_chain_points'] 
prep_indexed['storage_cap_lits']  = 1-prep_indexed['storage_cap_lits'] 
```

```python
dis = al['District_Key'].apply(lambda x : x[3:])
prep_indexed['district'] = dis
prep_indexed['state'] = al['state']
```

```python
prep_indexed.replace({0:1},inplace = True)
prep_indexed['index'] = np.power(prep_indexed['density']*prep_indexed['urbans']*prep_indexed['full_immunes']*prep_indexed['elderlies']*prep_indexed['hh_more_than_4']*prep_indexed['cold_chain_points']*prep_indexed['storage_cap_lits'],1/7)
prep_indexed.head()
```

```python
prep_indexed.to_csv('S:/Hackathons/EY/results/indexed1.csv',index = False)
```
